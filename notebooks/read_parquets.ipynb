```python
import pandas as pd
import numpy as np

kpis = pd.read_parquet('../data/processed/kpis.parquet')
ranked = pd.read_parquet('../data/processed/ranked.parquet')
```

```python
ranked.sort_values('median_vol_15d', ascending=True)
```

```python
ranked_clean = ranked.copy()

# Convert decimal ratios to percentages for better readability
ranked_clean['mom_7d_pct'] = np.round(ranked_clean['mom_7d'] * 100, 2)
ranked_clean['vol_7d_pct'] = np.round(ranked_clean['vol_7d'] * 100, 2)
ranked_clean['pct_positive_days_pct'] = np.round(ranked_clean['pct_positive_days'] * 100, 1)

# Round other columns to reasonable decimal places
ranked_clean['median_vol_15d_k'] = np.round(ranked_clean['median_vol_15d'] / 1000, 1)  # Convert to thousands
ranked_clean['potential_score_clean'] = np.round(ranked_clean['potential_score'], 3)

# Select and reorder columns for clean display
ranked_simplified = ranked_clean[[
    'rank', 'coin_id', 'symbol', 'name',
    'mom_7d_pct', 'vol_7d_pct', 'median_vol_15d_k', 
    'pct_positive_days_pct', 'potential_score_clean'
]].copy()

# Rename columns for clarity
ranked_simplified.columns = [
    'Rank', 'Coin ID', 'Symbol', 'Name',
    'Momentum 7D (%)', 'Volatility 7D (%)', 'Median Volume 15D (K)', 
    'Positive Days (%)', 'Potential Score'
]
```
